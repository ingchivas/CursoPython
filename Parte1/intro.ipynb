## Breve recorrido por la historia de Python

Python es el tercer lenguaje de programación más usado en el mundo.

Guido Van Rossum es el creador y responsable de que Python exista. Se trata de un informático de origen holandés que fue el encargado de diseñar Python y de pensar y definir todas las vías posibles de evolución de este popular lenguaje de programación

El nombre de este lenguaje de programación es en honor a los Monty Python, el famoso grupo de cómicos británicos.

En las navidades de 1989 Van Rossum, mientras trabajaba en un centro de investigación holandés (CWI), decidió empezar un nuevo proyecto como pasatiempo personal. Pensó en darle continuidad a ABC, un lenguaje de programación que se desarrollo en el mismo centro en el que estaba trabajando.

![EjABC](https://cdn-blog.adafruit.com/uploads/2020/01/adafruit_2019_3368.jpg)

La historia de Python empieza con Guido Van Rossum empezando su desarrollo en 1989 y empezando a implementarlo en febrero de 1991, momento en el que se publicó la primera versión pública: la 0.9.0.

La versión 1.0, que se publicó en enero de 1994, la versión 2.0 se publicó en octubre de 2000 y la versión 3.0 se publicó en diciembre de 2008.

Esta primera versión de Python ya incluía clases con herencias, manejo de excepciones, funciones y una de sus características fundamentales: funcionamiento modular. Esto permitía que fuese un lenguaje mucho más limpio y accesible para la gente con pocos conocimientos de programación. Una característica que se mantiene hasta el día de hoy.

````python
from tkinter import *

ventana = Tk()
#ventana.geometry("300x400")
ventana.title("Calculadora")

#Entrada de texto
e_texto = Entry(ventana, font = "Calibri 15")
e_texto.grid(row = 0, column = 0, columnspan = 4, padx = 5, pady = 5)      #Acomodada con grid y se indica que habrá 4 columnas después de esta

#Funciones
i = 0
def click_boton(valor):
    global i
    e_texto.insert(i,valor) #Con indice 0 para que se vayan acomodando los valores en la posición 
    i += 1

def borrar():
    global i
    #Agregar en un futuro la opción de borrar un solo caracter
    #e_texto.delete(i-1)   #Borrar desde el índice 0 hasta el final
    #i -= 1
    e_texto.delete(0,END)   #Borrar desde el índice 0 hasta el final
    i = 0

def evaluar():
    ecuacion = e_texto.get()
    resultado = eval(ecuacion)
    e_texto.delete(0,END)   #Borrar desde el índice 0 hasta el final
    e_texto.insert(0,resultado)
    i = 0

#Botones

boton1 = Button(ventana, text = "1", width = 5, height = 2, command = lambda: click_boton(1))
boton2 = Button(ventana, text = "2", width = 5, height = 2, command = lambda: click_boton(2))
boton3 = Button(ventana, text = "3", width = 5, height = 2, command = lambda: click_boton(3))
boton4 = Button(ventana, text = "4", width = 5, height = 2, command = lambda: click_boton(4))
boton5 = Button(ventana, text = "5", width = 5, height = 2, command = lambda: click_boton(5))
boton6 = Button(ventana, text = "6", width = 5, height = 2, command = lambda: click_boton(6))
boton7 = Button(ventana, text = "7", width = 5, height = 2, command = lambda: click_boton(7))
boton8 = Button(ventana, text = "8", width = 5, height = 2, command = lambda: click_boton(8))
boton9 = Button(ventana, text = "9", width = 5, height = 2, command = lambda: click_boton(9))
boton0 = Button(ventana, text = "0", width = 13, height = 2, command = lambda: click_boton(0))

boton_borrar = Button(ventana, text = "AC", width = 5, height = 2, command = borrar)
boton_parentesis1 = Button(ventana, text = "(", width = 5, height = 2, command = lambda: click_boton("("))
boton_parentesis2 = Button(ventana, text = ")", width = 5, height = 2, command = lambda: click_boton(")"))
boton_punto = Button(ventana, text = ".", width = 5, height = 2, command = lambda: click_boton("."))

boton_suma = Button(ventana, text = "+", width = 5, height = 2, command = lambda: click_boton("+"))
boton_resta = Button(ventana, text = "-", width = 5, height = 2, command = lambda: click_boton("-"))
boton_division = Button(ventana, text = "÷", width = 5, height = 2, command = lambda: click_boton("/"))
boton_mult = Button(ventana, text = "x", width = 5, height = 2, command = lambda: click_boton("*"))
boton_igual = Button(ventana, text = "=", width = 5, height = 2, command = evaluar)


#Posicionar botones

boton_borrar.grid(row = 1, column = 0, padx = 5, pady = 5)
boton_parentesis1.grid(row = 1, column = 1, padx = 5, pady = 5)
boton_parentesis2.grid(row = 1, column = 2, padx = 5, pady = 5)
boton_division.grid(row = 1, column = 3, padx = 5, pady = 5)

boton1.grid(row = 2, column = 0, padx = 5, pady = 5)
boton2.grid(row = 2, column = 1, padx = 5, pady = 5)
boton3.grid(row = 2, column = 2, padx = 5, pady = 5)
boton4.grid(row = 3, column = 0, padx = 5, pady = 5)
boton5.grid(row = 3, column = 1, padx = 5, pady = 5)
boton6.grid(row = 3, column = 2, padx = 5, pady = 5)
boton7.grid(row = 4, column = 0, padx = 5, pady = 5)
boton8.grid(row = 4, column = 1, padx = 5, pady = 5)
boton9.grid(row = 4, column = 2, padx = 5, pady = 5)
boton0.grid(row = 5, column = 0, padx = 5, pady = 5, columnspan = 2)

boton_mult.grid(row = 2, column = 3, padx = 5, pady = 5)
boton_suma.grid(row = 3, column = 3, padx = 5, pady = 5)
boton_resta.grid(row = 4, column = 3, padx = 5, pady = 5)
boton_igual.grid(row = 5, column = 3, padx = 5, pady = 5)
boton_punto.grid(row = 5, column = 2, padx = 5, pady = 5)


ventana.mainloop()
````

````python
import turtle as te
import numpy as np
import time as tm
#import os

#Creando mi tortuga
td = te.Turtle()
td.speed('fastest')
te.title('GATO 0234621')

#arreglo de numpy para llevar el control de las posiciones
tablerocmd = [['','',''],['','',''],['','','']]
tcli = np.array(tablerocmd)

#Arreglo np para identificar si la casilla ya fue utilizada
tps = [[False,False,False],[False,False,False],[False,False,False]]
tpos = np.array(tps)

def nclicks(c=[0]):#Contando el número de clicks que se dan, para determinar el turno de una manera sencilla
    
    c[0]+= 1
    
    return (c[0])


def tablero():
    td.up()
    td.color('white')
    td.goto(-48,188)
    td.down()
    td.write('GATO!!!!', font=('Comic Sans MS', 70))#Creo título
    te.bgcolor('black')#fondo negro, dark theme supremacy!
    td.pensize(2)
    
    td.up()
    td.ht()
    td.goto(0,0)
    td.down()
    #CREANDO EL CUADRADO
    td.forward(240)
    td.right(90)
    td.forward(240)
    td.right(90)
    td.forward(240)
    td.right(90)
    td.forward(240)

    #CREANDO LAS VERTICALES
    td.up()
    td.goto(80,-240)
    td.down()
    td.forward(240)
    
    td.up()
    td.goto(160,-240)
    td.down()
    td.forward(240)
    td.up()
    
    #CREANDO HORIZONTALES
    td.right(90)

    td.goto(0,-80)
    td.down()

    td.forward(240)
    td.up()

    td.goto(0,-160)
    td.down()
    td.forward(240)
    
def pos(x,y):#Función para determinar la posición del click, guardándola en vars globales para acceder a ella más tarde en px y py
    global px
    px = x
    global py
    py = y
    print("(X={0},Y={1})".format(x,y))#imprimo a la consola solo como referencia

def juego(x,y):#Le paso mis cordenadas
    global turno
    turno = nclicks()#Creo mi número de turno, que al exceder 9 y no haber ganador, termina el juego en empate
    pos(x,y)
    posvalid = True #Creo una bandera para ver si mi posición está dentro de los cuadra
    
    if 80 > px > 0 and -80 < py < 0:#Determino mi cuadrante viendo si las coordenadas del click entra en alguno de las 9 casillas
        #C1
        fil=0
        col=0

    elif 80 > px > 0 and -160 < py < -80:
        #C2
        fil=1
        col=0
        
    elif 80 > px > 0 and -240 < py < -160:
        #C3
        fil=2
        col=0
        
    elif 160 > px > 80 and -80 < py < 0:
        #C4
        fil=0
        col=1
        
    elif 160 > px > 80 and -160 < py < -80:
        #C5
        fil=1
        col=1
        
    elif 160 > px > 80 and -240 < py < -160:
        #C6
        fil=2
        col=1
        
    elif 240 > px > 160 and -80 < py < 0:
        #C7
        fil=0
        col=2
        
    elif 240 > px > 160 and -160 < py < -80:
        #C8
        fil=1
        col=2
        
    elif 240 > px > 160 and -240 < py < -160:
        #C9
        fil=2
        col=2
        
    else:
        print("Posición fuera de tablero")#En caso de que no se encuentre en uno de los cuadrantes 
        posvalid = False #Si mi posición es inválida, me avisa y no me permite colocar el símbolo
        
    if ((turno%2==0) and (posvalid == True)and (tpos[fil][col]==False)):#Valido si mi turno es par [X], si mi posición es válida, y si mi casilla no está ocupada
        td.color('red')#Color rojo para la espantosa X
        td.pensize(3)
        td.up()
        td.goto(0+col*80,0-fil*80)#Primera lateral
        td.down()

        td.right(45)
        td.forward(113)
        td.right(90)

        td.up()
        td.goto(80+col*80,0-fil*80)#Segunda lateral
        td.down()

        td.forward(113)
        td.left(135)

        tcli[fil][col]="X"#Coloco una X en mi almacenador de posiciones
        tpos[fil][col]=True#Le digo que mi posición ya está ocupada
        turno += 1#Paso al sig. turno
        ganar()#Checo si mi usuario ganó 
        print(tcli)#Imprimo mi arreglo de np

    elif ((turno%2 != 0) and (posvalid == True)and (tpos[fil][col]==False)):#Lo mismo que arriba pero para turnos impares [O]
        td.color('green')
        td.pensize(3)
        td.up()
        td.goto(40+col*80,-80-fil*80)
        td.down()
        td.circle(40)#En este caso es más fácil, solo hago un círculo
        
        tcli[fil][col] = "O"
        tpos[fil][col]=True
        turno += 1
        ganar()
        print(tcli)
        
def ganax():#Funcion si es que gana [X]
    td.up()
    td.goto(0,0)
    td.down()
    td.color('blue')
    td.write('Gana la espantosa [X]', font=('Comic Sans MS', 18))

def ganao():#Función si es que gana [O]
    td.up()
    td.goto(0,0)
    td.down()
    td.color('blue')
    td.write('Gana [O]', font=('Comic Sans MS', 18))

def H1():#Trazo la línea horizontal fila 0, en caso de que se gane de esta forma
    td.color('lawn green')
    td.up()
    td.goto((0,-37))
    td.down()
    td.goto(240,-37)

def H2():#Trazo la línea horizontal fila 1, en caso de que se gane de esta forma
    td.color('lawn green')
    td.up()
    td.goto(0,-121)
    td.down()
    td.goto(240,-121)

def H3():#Trazo la línea horizontal fila 2, en caso de que se gane de esta forma
    td.color('lawn green')
    td.up()
    td.goto((0,-195))
    td.down()
    td.goto(240,-195)

def V1():#Vertical Columna 0
    td.color('lawn green')
    td.up()
    td.goto((36,-0))
    td.down()
    td.goto(36,-240)

def V2():#Vertical Columna 1
    td.color('lawn green')
    td.up()
    td.goto((120,0))
    td.down()
    td.goto(120,-240)

def V3():#Vertical Columna 1
    td.color('lawn green')
    td.up()
    td.goto((200,0))
    td.down()
    td.goto(200,-240)

def D1():#Diag Izquierda
    td.color('lawn green')
    td.up()
    td.goto((0,-0))
    td.down()
    td.goto(240,-240)

def D2():#Diag Derecha
    td.color('lawn green')
    td.up()
    td.goto((240,0))
    td.down()
    td.goto(0,-240)



def ganar():
    #Función para determinar el ganador, dependiendo de las diferentes posiciones de los símbolos en el arreglo de numpy
    ganador = False #Flag para checar el estado del ganador
    #PARA X
    if tcli[0][0]=='X' and tcli[0][1]=='X' and tcli[0][2]=='X':#H1Horizontal1
        H1()#Llamo la función que dibuja la línea que marca el ganador dependiendo de que forma gane
        ganax()
        ganador = True #Levanto flag
        tm.sleep(2) #Doy 2segundos para ver resultado
        te.bye()#Cierro programa
    
    elif tcli[1][0]=='X' and tcli[1][1]=='X' and tcli[1][2]=='X':#H2Horizontal2
        H2()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[2][0]=='X' and tcli[2][1]=='X' and tcli[2][2]=='X':#H3Horizontal3
        H3()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[0][0]=='X' and tcli[1][1]=='X' and tcli[2][2]=='X':#DIAGIZQ
        D1()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[2][0]=='X' and tcli[1][1]=='X' and tcli[0][2]=='X':#DIAGDER
        D2()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[0][0]=='X' and tcli[1][0]=='X' and tcli[2][0]=='X':#V1
        V1()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[0][1]=='X' and tcli[1][1]=='X' and tcli[2][1]=='X':#V2
        V2()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[0][2]=='X' and tcli[1][2]=='X' and tcli[2][2]=='X':#V3
        V3()
        ganax()
        ganador = True
        tm.sleep(2)
        te.bye()

    # Lo mismo pero PARA [O]
    elif tcli[2][0]=='O' and tcli[1][1]=='O' and tcli[0][2]=='O':#DIAGDER
        D2()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[0][0]=='O' and tcli[1][1]=='O' and tcli[2][2]=='O':#DIAGIZQ
        D1()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()
   
    elif tcli[0][0]=='O' and tcli[0][1]=='O' and tcli[0][2]=='O':#H1
        H1()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()

    elif tcli[1][0]=='O' and tcli[1][1]=='O' and tcli[1][2]=='O':#H2
        H2()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()

    elif tcli[2][0]=='O' and tcli[2][1]=='O' and tcli[2][2]=='O':#H3
        H3()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()
    

    elif tcli[0][0]=='O' and tcli[1][0]=='O' and tcli[2][0]=='O':#VERT1
        V1()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()
    
    elif tcli[0][1]=='O' and tcli[1][1]=='O' and tcli[2][1]=='O':#VERT2
        V2()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()

    elif tcli[0][2]=='O' and tcli[1][2]=='O' and tcli[2][2]=='O':#VERT3
        V3()
        ganao()
        ganador = True
        tm.sleep(2)
        te.bye()

    elif (turno == 10 and ganador == False): #En caso de empate, checo turno y si existe ganador
        td.up()
        td.goto(0,0)
        td.down()
        td.color('blue')
        td.write('EMPATE!!!!', font=('Comic Sans MS', 18))#Escribo empate
        tm.sleep(2)
        te.bye()#Cierro turtle
    
    
    else:
        pass#En caso de que ninguna condición de cumpla, sigo hasta que se genere



def jugar():#Llamo al generador para crear mi juego de gato
    td.speed('fastest')
    te.title('GATO 0234621')
    te.Screen().onscreenclick(juego,1,add=True)
    te.onscreenclick(pos,1,add=True)
    tablero()
    te.done()

jugar()
````
